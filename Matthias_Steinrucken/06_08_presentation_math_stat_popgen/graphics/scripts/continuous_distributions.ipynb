```python
import numpy
import scipy
import matplotlib.pyplot as plt
import matplotlib
```

#### uniform

```python
# get probabilities
xs = numpy.linspace (-1,5,2000)
ys = numpy.zeros (len(xs))
ys[(xs>1) & (xs<3)] = 0.5


# text size
matplotlib.rcParams["font.size"] = 22
(fig, ax) = plt.subplots (figsize=(9,5))

# plot them
ax.plot (xs, ys, label="Unif[1,3]", lw=4)
ax.fill_between (xs, ys, alpha=0.1)
ax.set_ylim ((0,0.7))
ax.set_xlim ((-0.5,4.5))
ax.set_ylabel ("f(x)")
ax.set_xlabel ("x")
ax.set_title ("Uniform distribution")
ax.legend()

# save plot
plt.tight_layout()
plt.savefig ("uniform_density.pdf")
```

#### normal distribution

```python
# prob params
normParams = [
    (0, 1),
    (0, 0.5),
    (2, 1),
]

# plot params
xLim = (-3,7)
yLim = (0,1)
xs = numpy.linspace (xLim[0], xLim[1], 2000)

# text size
matplotlib.rcParams["font.size"] = 22
(fig, ax) = plt.subplots (figsize=(9,5))

# go though params
for (pIdx, thisP) in enumerate(normParams):
    # get values of density
    (thisMu, thisSigma) = thisP
    ys = scipy.stats.norm.pdf (xs, loc=thisMu, scale=thisSigma)

    # plot them
    ax.plot (xs, ys, label=fr"$\mu$ = {thisMu:.1f}, $\sigma$ = {thisSigma:.1f}", lw=4)
    ax.fill_between (xs, ys, alpha=0.1)

ax.set_ylim (yLim)
ax.set_xlim (xLim)
ax.set_ylabel ("f(x)")
ax.set_xlabel ("x")
ax.set_title ("Normal distribution")
ax.legend()

# save plot
plt.tight_layout()
plt.savefig ("normal_distribution.pdf")
```

#### exponential distribution

```python
# prob params
lambdas = [0.1, 0.2]

# plot params
xLim = (0,30)
yLim = (0,0.25)
xs = numpy.linspace (xLim[0], xLim[1], 2000)

# text size
matplotlib.rcParams["font.size"] = 22
(fig, ax) = plt.subplots (figsize=(9,5))

# go though params
for (lIdx, thisL) in enumerate(lambdas):
    # get values of density
    ys = scipy.stats.expon.pdf (xs, scale=1/thisL)

    # plot them
    ax.plot (xs, ys, label=fr"$\lambda$ = {thisL:.1f}", lw=4)
    ax.fill_between (xs, ys, alpha=0.1)

ax.set_ylim (yLim)
ax.set_xlim (xLim)
ax.set_ylabel ("f(x)")
ax.set_xlabel ("x")
ax.set_title ("Exponential distribution")
ax.legend()

# save plot
plt.tight_layout()
plt.savefig ("exponential_distribution.pdf")
```

#### beta distribution

```python
# prob params
betaParams = [
    (0.1, 0.1),
    (1, 1),
    (10, 10),
    (5, 10),
]

# plot params
xLim = (0,1)
yLim = (0,7.5)
EPSILON = 1e-3
xs = numpy.linspace (xLim[0]+EPSILON, xLim[1]-EPSILON, 1000)

# text size
matplotlib.rcParams["font.size"] = 22
(fig, ax) = plt.subplots (figsize=(7,6))

# go though params
for (bIdx, thisB) in enumerate(betaParams):
    # get values of density
    (thisAlpha, thisBeta) = thisB
    ys = scipy.stats.beta.pdf (xs, a=thisAlpha, b=thisBeta)

    # plot them
    ax.plot (xs, ys, label=fr"$(\alpha, \beta)$ = ({thisAlpha:.1f}, {thisBeta:.1f})", lw=4)
    ax.fill_between (xs, ys, alpha=0.1)

ax.set_ylim (yLim)
ax.set_xlim (xLim)
ax.set_ylabel ("f(x)")
ax.set_xlabel ("x")
ax.set_title ("Exponential distribution")
ax.legend()

# save plot
plt.tight_layout()
plt.savefig ("beta_distribution.pdf")
```
