```python
import numpy
import matplotlib.pyplot as plt
```

```python
Ne = 10000
s = 0.01
```

```python
def p (k,s,pZero):
    numerator = numpy.power (1+s, k) * pZero
    denominator = (1 - pZero) + numerator
    return numerator / denominator
```

```python
def pLog (t,s,pZero):
    numerator = numpy.exp (s*t) * pZero
    denominator = (1 - pZero) + numerator
    return numerator / denominator
```

```python
ks = numpy.arange (0,2000)
# ss = [s/2, s, 2*s]
ss = [s]
for thisS in ss:
    plt.plot (ks, p (ks, thisS, 1/(2*Ne)))
    # plt.plot (ks, p (ks/20, 20*thisS, 20/(2*Ne)))

# cosmetics
plt.ylabel ("p")
plt.xlabel ("generation")
# plt.title ("Initial Frequency 1/(2Ne) with Ne = 10,000")
plt.title ("p(t)")

# legend
theLegend = []
for thisS in ss:
    theLegend.append (f"s = {thisS:.2f}")
plt.legend (theLegend, loc="lower right")

# save it
plt.savefig ("selection_trajectory.pdf")
```
