```python
import numpy
import scipy
import matplotlib.pyplot as plt
import matplotlib
```

#### fake log-likelihood

```python
# get probabilities
xs = numpy.linspace (-1,5,2000)
ys = scipy.stats.norm.pdf (xs, loc=2.5, scale=0.1)
ys = numpy.log10 (ys)
ys -= 4500

# text size
matplotlib.rcParams["font.size"] = 22
(fig, ax) = plt.subplots (figsize=(9,5))

# plot them
ax.plot (xs, ys, lw=4)
# ax.fill_between (xs, ys, alpha=0.1)
# ax.set_ylim ((0,0.7))
# ax.set_xlim ((-0.5,4.5))
ax.set_ylabel (r"$L(\theta)$")
ax.set_xlabel (r"$\theta$")
ax.set_title ("Log-likelihood")
ax.axvline (2.5, ls='--', lw=4, color="C1")
# ax.legend()

# save plot
plt.tight_layout()
plt.savefig ("log_likelihood.pdf")
```
